```python
#Q7
print(df.isnull().sum())
print(df.isnull().sum().sum())
```

```python
#Q8

df['serious_or_critical'].fillna(0,inplace=True)

sum(df['serious_or_critical'].isnull())

df['total_deaths'].fillna(0,inplace=True)

sum(df['total_deaths'].isnull())

df['total_recovered'].fillna(0,inplace=True)

sum(df['total_recovered'].isnull())

df['active_cases'].fillna(0,inplace=True)

sum(df['active_cases'].isnull())

df['total_deaths_per_1m_population'].fillna(0,inplace=True)

sum(df['total_deaths_per_1m_population'].isnull())

df['total_tests'].fillna(0,inplace=True)

sum(df['total_tests'].isnull())

df['total_tests_per_1m_population'].fillna(0,inplace=True)

sum(df['total_tests_per_1m_population'].isnull())


#Checking for null values again
print(df.isnull().sum())
print(df.isnull().sum().sum())
```

```python
#Q10
import matplotlib.pyplot as plt
df_continent = df.groupby(['continent']).sum().reset_index()
display(df_continent)
plt.pie(df_continent['total_confirmed'], labels = df_continent['continent'], radius = 1.5, autopct = '%0.2f%%')
plt.show()
```
